# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style

import os
```

#Import CSV File

```python
# Read the data from the Excel file
df = pd.read_csv('Top 10 (Most Runs).csv')
```

#CSV to List Conversion

```python
Player = df['Player'].tolist()
Matches = df['Mat'].tolist()
Innings = df['Inns'].tolist()
BF = df['BF'].tolist()
Runs = df['Runs'].tolist()
Ducks = df['0'].tolist()
Fours = df['4s'].tolist()
Sixes = df['6s'].tolist()
Fifty = df['50'].tolist()
HS = df['HS'].tolist()
Avg = df['Ave'].tolist()
SR = df['SR'].tolist()
```

#Plot simple Bar Graph

```python
plt.bar(Player, Runs)
```

#Resizing the Bar plot

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs)
```

#Bar Colouring

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs, color='g')
```

##Adjust column / bar transparency

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs, color='g', alpha=0.5)
```

##Adjust multiple colorus

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs, color=['g','r','b','y','m'])
```

#Bar Edge Designing

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs, edgecolor='black')
```

##Adjust edge colour & thickness

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs, edgecolor='black', linewidth=5)
```

##Adjust edge style

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs, edgecolor='black', linewidth=5, linestyle=":")
```

#Title & Labels customization

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs)

plt.xlabel('Players')
plt.ylabel('Runs')
plt.title("ICC Men's T20 World Cup 2024 - Top Run Scorer")
```

##Adjust label's font size

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs)

plt.xlabel('Players', fontsize=10)
plt.ylabel('Runs', fontsize=10)
plt.title("ICC Men's T20 World Cup 2024 - Top Run Scorer", fontsize=15)
```

##Adjust label's font style

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs)

plt.xlabel('Players', fontsize=10, fontstyle='italic')
plt.ylabel('Runs', fontsize=10, fontstyle='italic')
plt.title("ICC Men's T20 World Cup 2024 - Top Run Scorer", fontsize=15, fontstyle='oblique')
```

##Adjust label's colour

```python
fig=plt.figure(figsize=(24,6))
plt.bar(Player, Runs)

plt.xlabel('Players', fontsize=10, fontstyle='oblique')
plt.ylabel('Runs', fontsize=10, fontstyle='oblique')
plt.title("ICC Men's T20 World Cup 2024 - Top Run Scorer", fontsize=15, color='g')
```

#Double Bar Graph

```python
xpos = np.arange(len(Player))
xpos
```

```python
style.use('ggplot')
plt.figure(figsize=(24,6))
barWidth = 0.2

plt.bar(xpos, BF, color='royalblue', width = barWidth, label='Balls Faced')
plt.bar(xpos+0.2, Runs, color='red', width = barWidth, label='Runs Scored')

plt.xticks(xpos+0.3, df['Player'])

plt.xlabel('Players')
plt.ylabel('Balls Faced VS Runs Scored')
plt.title("ICC Men's T20 World Cup 2024 - Balls Faced VS Runs Scored")
plt.legend()
```

#Stacked Bar Chart / Graph

```python
x = df['Player']
xpos = np.arange(len(x))
barWidth = 0.4

plt.figure(figsize=(24,6))
plt.bar(xpos, BF, color='royalblue', width= barWidth, label='Balla Faced')
plt.bar(xpos, Runs, bottom=BF, color='red', width= barWidth, label='Runs Scored')

plt.xlabel('Players')
plt.ylabel('Balls Faced VS Runs Scored')
plt.title("ICC Men's T20 World Cup 2024 - Balls Faced VS Runs Scored")
plt.xticks(xpos, x)
plt.legend()
```

#Horizontal Stacked Bar Chart / Graph

```python
fig = plt.figure(figsize=(24,10))

plt.barh(Player, BF, color='blue', height=0.5, label='Balls Faced')
plt.barh(Player, Runs, left=BF, color='red', height=0.5, label='Runs Scored')
plt.ylabel('Player')
plt.xlabel('Balls Faced VS Runs Scored')
plt.title("ICC Men's T20 World Cup 2024 - Balls Faced VS Runs Scored")
plt.legend()
plt.show()
```
